##Задача 4.3
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
plt.figure(figsize=(10, 8))
sns.set(font_scale=1.4)
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm')
plt.title('Correlation matrix');
```

```python
sns.jointplot(x=df['price'], y=df['sqft_living'], kind='reg');
```

C увеличением жилой площади дома растёт и его цена, однако здесь нет линейной зависимости, возможно на цену дома оказывают влияние его состояние и месторасположение, на графике видно, что цена за дома в 4000 квадратов меняется от 0,3 до 3,3 единиц измерения.

```python
sns.jointplot(x=df['price'], y=df['grade'], kind='reg');
```

Настоящий график подтверждает сделанные выше выводы о зависимости цены дома от его состояния, но и в данном случае при видимой линейности линии регрессии в глаза бросается существенный разброс цен на дома аналогичного состояния.

```python
sns.boxplot(x=df['price'], y=df['grade'].astype('str'), whis=1.5);
plt.xlabel('price')
plt.ylabel('grade')
plt.title('Зависимость цены от состояния дома');
```

Из данного графика видно, что наибольшей цены достигают дома уровня "13", при этом есть дом, "11" уровня, который стоит больше домов уровней "12" и "13" вероятно его цена обусловлена другими факторами.

```python
sns.boxplot(x=df['price'], y=df['floors'].astype('str'), whis=1.5);
plt.xlabel('price')
plt.ylabel('floors')
plt.title('Зависимость цены от этажности дома');
```

Данный график чётко показывает, что именно двухэтажные дома находятся в зоне цен выше средних и хотя общая масса домов в 2,5 этажа стоят больше основной массы 2-ух этажных строений, учитывая количество тех и других домов 38% против 0,7% можно утверждать, что законадателем цен являются двухэтажные дома.

```python
sns.boxplot(x=df['price'], y=df['view'].astype('str'), whis=1.5);
plt.xlabel('price')
plt.ylabel('view')
plt.title('Зависимость цены от месторасположения.(красивый вид)');
```

Можно предположить, что вид соответствующий цифре "4" является наиболее благовидным, так как дома обладающие этим видом из окна в подавляющей массе стоят дороже всех остальных.
